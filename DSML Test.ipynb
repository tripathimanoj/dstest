```python
car["x_4_year"].fillna(value=np.mean(car["x_4_year"]),inplace=True)      # filling null value with mean of that col.
car["Price in thousands"].fillna(value=np.mean(car["Price in thousands"]),inplace=True)
car["Engine size"].fillna(value=np.mean(car["Engine size"]),inplace=True)
car["Horsepower"].fillna(value=np.mean(car["Horsepower"]),inplace=True)
car["Wheelbase"].fillna(value=np.mean(car["Wheelbase"]),inplace=True)
car["Width"].fillna(value=np.mean(car["Width"]),inplace=True)
car["Length"].fillna(value=np.mean(car["Length"]),inplace=True)
car["Curb weight"].fillna(value=np.mean(car["Curb weight"]),inplace=True)
car["Fuel capacity"].fillna(value=np.mean(car["Fuel capacity"]),inplace=True)
car["Fuel efficiency"].fillna(value=np.mean(car["Fuel efficiency"]),inplace=True)
car["XX"].fillna(value=np.mean(car["XX"]),inplace=True)

car.isnull().sum()
```
